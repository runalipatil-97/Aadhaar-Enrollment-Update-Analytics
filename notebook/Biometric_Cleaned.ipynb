```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
path = r"C:\Runali\Hackathon\api_data_aadhar_biometric\*.csv"
all_files = glob.glob(path)

all_files
```

```python
df_list = []

for file in all_files:
    df = pd.read_csv(file)
    df_list.append(df)

df_all = pd.concat(df_list, ignore_index=True)

df_all.shape
```
